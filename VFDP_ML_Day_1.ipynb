```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=4, input_shape=[1], activation=None),# hidden layer
                            keras.layers.Dense(units=1)])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs,ys, epochs = 100)

```

```python
ex = np.array([5.0],dtype=float)
model.predict(ex)

```

```python
ex = np.array([10.0],dtype=float)
model.predict(ex)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=2, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mae'])

xs = np.array([2,4,6,8,10], dtype=float)
ys = np.array([[8.5,12],
               [10.0,16],
               [11.5,20],
               [13.0,26],
               [14.0,32]], dtype=float)

model.fit(xs,ys, epochs = 500)

```

```python
ex = np.array([7.0],dtype=float)
model.predict(ex)
```

```python
ex = np.array([10.0],dtype=float)
model.predict(ex)
```
